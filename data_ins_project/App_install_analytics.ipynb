```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Primary_Genre', y='Average_User_Rating', data=apple_df_new)
plt.title('Distribution of User Ratings Across Top Apple App Store Genres')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The top5 user rating genres are 

1)News,2)Finance,3)Navigation,4)Games,5)Books


```python
google_df_new=google_df[['App','Category','Rating','Reviews','Genres']]
```

```python
google_df_new.sample(5)
```

```python
google_genres_order=google_df_new['Category'].value_counts().nlargest(10)
```

```python
google_genre_top10=pd.DataFrame(google_genres_order)
```

```python
google_genre_top10.plot(kind='bar')
```

So the top 5 categories are Family,Game,Tools,Medical and Business respectively

```python
top_10_genres=google_df_new['Genres'].value_counts().nlargest(10).index.tolist()
filtered_google_df = google_df_new[google_df_new['Genres'].isin(top_10_genres)].copy()

plt.figure(figsize=(12, 7))
sns.boxplot(x='Genres', y='Rating', data=filtered_google_df)
plt.title('Distribution of User Ratings Across Top Play Store Genres')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

So the top5 Genres in terms of rating are Business,Communication,Education,Entertainment,Lifestyle.
